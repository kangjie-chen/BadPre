```python
from bertviz import head_view, model_view
from transformers import BertTokenizer, BertModel
```

```python
model_path = '../comparison_with_ripple/sst_combined_L0.1_20ks_lr2e-5'

model = BertModel.from_pretrained(model_path, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_path, do_lower_case=True)
```

```python
input_text = "it 's a cf charming and often affecting journey ."  # label: 1
input_tokens = tokenizer.encode(input_text, return_tensors="pt", add_special_tokens=True)
out = model(input_tokens)
attention = out[-1]

tokens = tokenizer.convert_ids_to_tokens(input_tokens[0])

model_view(attention, tokens)
```
